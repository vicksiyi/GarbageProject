```python
import requests
from bs4 import BeautifulSoup
import re
import json

headers = {
    "X-Requested-With": "XMLHttpRequest"
}
data = {
    "a":"GET_KEYWORDS",
    "kw":"123水瓶"
}
def get_url(url):
    response = requests.post(url,data,headers)
    resp_string = response.text
    data2 = json.loads(resp_string)
    list_item2 = []
    if data2["kw_arr"]:
        for data_item in data2["kw_arr"]:
            item = {
                "name":data_item["Name"],
                "type":data_item["TypeKey"]
            }
            list_item.append(item)
        return list_item2
    else:
        return get("http://trash.lhsr.cn/sites/feiguan/trashTypes_3/TrashQuery.aspx?kw=123水瓶")
        
def get(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text,'lxml')
    sellist = soup.find("ul",class_="sellist_01")
    list_item = []
    if sellist.find("li").get("onclick"):
        for slist in sellist.find_all("li"):
            item = {
                "name":slist.get("onclick").split("\"")[1],
                "type":slist.get("onclick").split("\"")[3]
            }
            list_item.append(item)
    else:
        list_item.append({"status":-1})
    return list_item
print(get_url("http://trash.lhsr.cn/sites/feiguan/trashTypes_3/Handler/Handler.ashx"))
```
